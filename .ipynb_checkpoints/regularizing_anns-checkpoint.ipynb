```python
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt

import keras
from keras.layers import Dense, Dropout, Flatten
from keras.models import Sequential

from keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_test
```

```python
plt.imshow(x_train[0])
```

```python
x_train[0].shape
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(256, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
from keras.optimizers import Adam
```

```python
Adam()
```

```python
#training
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs = 200, verbose = 0)
```

```python
plt.plot(history.history['acc'], '--o', label = 'train')
plt.plot(history.history['val_acc'], '--o', label = 'validation')
plt.legend();
```

### Regularization I: EarlyStopping

```python
from keras.callbacks import EarlyStopping
```

```python
stopper = EarlyStopping(patience = 3)
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(256, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
          callbacks=stopper, epochs = 200)
```

### Regularization II: `l1` and `l2` regularizers

```python
from keras.regularizers import l1, l2, l1_l2
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(256, activation = 'relu', kernel_regularizer = l2(0.001)))
#                bias_regularizer = 'l2',
#                activity_regularizer = 'l2'))
model.add(Dense(10, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
          callbacks=stopper, epochs = 10)
```

### Regularization III: Dropout

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(256, activation = 'relu', kernel_regularizer = l2(0.001)))
model.add(Dropout(0.1))
model.add(Dense(10, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
          callbacks=stopper)
```

### Using with `sklearn`

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
def network_maker(dropout_amt = 0.1):
    model = Sequential()
    model.add(Flatten())
    model.add(Dense(256, activation = 'relu', kernel_regularizer = l2(0.001)))
    model.add(Dropout(dropout_amt))
    model.add(Dense(10, activation = 'softmax'))
    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
    return model
```

```python
kclass = KerasClassifier(build_fn=network_maker)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(kclass, param_grid={'dropout_amt': [0.1, 0.8]})
```

```python
grid.fit(x_train, y_train)
```

```python
kclass = KerasClassifier(build_fn=network_maker)
sscaler = StandardScaler()
```

```python
class Account:
    limit = -1000
    class_name = 'Account'
    
    def __init__(self, name, amount):
        self.name = name
        self.amount = amount
        self.overdrawn_ = None
        
    def __str__(self):
        return f'{self.class_name}, {self.name}, {self.amount}'
        
    def withdraw(self, howmuch):
        new_amount = self.amount - howmuch
        if new_amount < self.limit:
            print("Sorry, that is over your limit!")
            self.overdrawn_ = True
        else:
            self.amount = new_amount
            
    def compare_rates_(account_1, account_2):
        return account_1 > account_2
            
    def deposit(self, howmuch):
        a = compare_rates_(a, b)
        self.amount += howmuch
        
    def statement(self):
        return f'Owner: {self.name}, Amount: {self.amount}'
    
    def testing_self(self, a, b):
        return self.amount + a - b
    
```

```python
class SavingAccount(Account):
    interest_rate = 0.25
    
    def add_interest(self):
        self.deposit(self.amount * self.interest_rate)
        
    def withdraw(self, howmuch):
        if self.amount - howmuch < 0:
            print('Savings cannot be overdrawn.')
        else:
            super(SavingAccount, self).withdraw(howmuch)
        
```

```python
jacob_savings = SavingAccount('Jacob', 10)
```

```python
jacob_savings.testing_self(4, 5)
```

```python
print(jacob_savings)
```
